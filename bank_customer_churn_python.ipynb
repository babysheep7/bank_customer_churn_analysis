```python
import pandas as pd

df = pd.read_csv('Bank_Churn.csv')
df.head()
```

```python
df.info
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
# Overall churn rate
churn_rate = df['Exited'].mean()
print(f'Overall churn rate: {churn_rate:.2%}')
print(f'Churned customers: {df["Exited"].sum():,} / {len(df):,}')
```

```python
# Rename columns to snake_case
df.columns = df.columns.str.replace(r'(?<!^)(?=[A-Z])', '_', regex=True).str.lower()
df.columns
```

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[18, 30, 40, 50, 60, 100],
    labels=['18-30', '31-40', '41-50', '51-60', '60+'],
    include_lowest=True
)
df[['age', 'age_group']].head(10)
```

```python
df['balance_segment'] = pd.cut(
    df['balance'],
    bins=[-1, 0, 75000, 150000, float('inf')],
    labels=['Zero', 'Low', 'Medium', 'High']
)
df[['balance', 'balance_segment']].head(10)
```

```python
df['credit_tier'] = pd.cut(
    df['credit_score'],
    bins=[0, 499, 599, 699, 850],
    labels=['Poor', 'Fair', 'Good', 'Excellent']
)
df[['credit_score', 'credit_tier']].head(10)
```

```python
df['tenure_segment'] = pd.cut(
    df['tenure'],
    bins=[-1, 2, 6, 100],
    labels=['New (0-2yr)', 'Regular (3-6yr)', 'Loyal (7+yr)']
)
df[['tenure', 'tenure_segment']].head(8)
```

```python
print('Final shape:', df.shape)
print('\nColumns:', df.columns.tolist())
print('\nMissing values:')
print(df.isnull().sum())
df.head()
```
